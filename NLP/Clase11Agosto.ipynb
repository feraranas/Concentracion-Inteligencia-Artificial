```python
import librosa
import soundfile as sf
import numpy as np
from IPython.display import Audio
```

### 1. Choose a sound in .wav

```python
# Loading the audio as wave form 'y',
# Storing the sampling rate as 'sr'
y, sr = librosa.load('bear_growl_y.wav')
```

```python
# Why is it 99,962? What does it mean to a wave form to be represented as an array?
y.mean
```

### 2. Plot the wavesound

```python
import librosa.display
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
librosa.display.waveshow(y, sr = sr)
```

### 3. Reduce the sound duration in seconds to half the time

```python
# Calculate the new target sampling rate to reduce duration by half
new_sr = int(sr * 2)
```

```python
# Resample the audio to the new sampling rate
resampled_audio = librosa.resample(y, orig_sr=sr, target_sr=new_sr)
```

```python
# Save the resampled audio to a new file
output_file = "resampled_bear_audio.wav"
sf.write(output_file, resampled_audio, new_sr)
```

### 4. Calculate the highest and lowest value of the wave amplitude

```python
y, sr = librosa.load('resampled_bear_audio.wav')

# Calculate the highest and lowest values of the wave amplitude
max_amplitude = np.max(y)
min_amplitude = np.min(y)

print("Highest amplitude:", max_amplitude)
print("Lowest amplitude:", min_amplitude)
```

### 5. Choose a new sound or song and plot both wave sounds in the same display to compare them in two different colors and transparency

```python
# New sound: included audio example
filename = librosa.example('nutcracker')

y2, sr2 = librosa.load(filename)

plt.figure(figsize=(12,4))
librosa.display.waveshow(y, sr = sr)
librosa.display.waveshow(y2, sr = sr2)
```
