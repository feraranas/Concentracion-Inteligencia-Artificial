````python
# NumPy is a fundamental package for numerical computations in Python. 

# It provides support for arrays (including multi-dimensional arrays) and a variety of 
# mathematical functions that are essential for scientific and data analysis tasks.

# NumPy's arrays (often referred to as NumPy arrays or ndarrays) are highly efficient 
# and provide a lot of functionality for numerical operations. For example, you can use 
# NumPy arrays to perform element-wise mathematical operations, linear algebra operations,
# and more.

import numpy as np

# Generate random data using NumPy
np.random.seed(42)  # Set a seed for reproducibility
x = np.random.rand(50)  # 50 random values between 0 and 1
y = 2 * x + 1 + np.random.randn(50) * 0.2  # Linear relationship with some noise

# Create a scatter plot using Seaborn
sns.scatterplot(x=x, y=y)

# Show the plot
plt.show()
````

````python
# The np.zeros function creates an array filled with zeros. The argument you pass 
# to the function specifies the shape of the array you want to create.

# In this case, you've passed 2 as the argument, so a will be a 1-dimensional NumPy array
# with a shape of (2,), containing two elements, both set to 0.
a = np.zeros(2)
````

# Importar Archivos Locales

Si deseamos cargar un archivo desde nuestra computadora, simplemente usamos:
```
from google.colab import files
uploaded = files.upload()
```

Muestra un diccionario de los archivos que se carcaron. El diccionario está codificado según el nombre del archivo y los valores con los datos que se subieron


````python
from google.colab import files
uploaded = files.upload()
````

````python
uploaded.keys()
````

````python
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
````
