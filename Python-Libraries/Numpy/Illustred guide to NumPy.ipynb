## Measures of Error

MeanSquaredError = $\sum_{i=1}^{n}(\hat{y}_{i}-y_{i}) ^2
$
You can use the functions ```np.square```  

````python
# Compute the MeanSqureError
````

````python
#Write your code here
error = (np.sum( np.square(data[:,6] - y_pred) )) / y.shape[0]
error2 = np.mean( np.square(y_pred - data[:,6]) )
````

````python
print(error)
print(error2)
````
