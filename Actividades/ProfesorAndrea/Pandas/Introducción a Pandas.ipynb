# Estructuras de Datos de Pandas

## Series

 Las series son *arreglos* de una dimensión etiquetados capaces de almacenar cualquier tipo de datos (integer, string, floar , objetos de python).

 Podemos entender a una Serie como una columna de una hoja de excel.
 Las etiquetas no deben ser necesariamente strings, pero deben ser objetos "hashable".

 Las operaciones entre Series (+,-,*,/,**) se basa entre el valor indexado, se permite que sean de diferentes tamaños.

 Para consultar los atributos y métodos propios del objeto Serie consulta la documentación oficial en: [Series](https://pandas.pydata.org/docs/reference/api/pandas.Series.html).


````python
#Construcción desde lista
data = [10,20,30,40]

ser = pd.Series(data)
print(ser)
````

````python
# Construcción desde diccionario

d = {'a':1,'b':2,'c':3}
d
````

````python
serDic=pd.Series(d, index=['x','y','z'])
print(serDic)
````

¿Qué pasó?!!!!!

Las llaves del diccionario deben coicidir con los valores de los índices.

````python
ser_dic = pd.Series(d,index=['a','b','c'])
print(ser_dic)
````

````python
# Acceder a elemento en la Serie por etiqueta o por número de índice

ser_dic['a']
````

````python
ser_dic[0]
````

#Selección basada en índices

Pandas tiene sus propios operadores  `loc` y `iloc`, para regresar partes del DF.

El orden es  Fila, Columna

````python
df.iloc[0]
````

Regresar la 1er columna:
El operador regresa todas las filas, y la primer columna

````python
df.iloc[:,0]
````

````python
df.iloc[:3,0]
````

Podemos pasar una lista

````python
df.iloc[[1,3,5,7],0]
````

¿Qué pasa si es un número negativo?

````python
df.iloc[-2:,:]
````

# Selección basada en etiquetas
 Para este paradigma las etiquetas es sobre los valores index, no su posición.


````python
df.loc[0,"variety"]
````

````python
df.loc[:5,["variety","petal.length"]]
````

````python
df.loc[:3,"sepal.length"]
````

#### ¡Cuidado!
`iloc` usa el esquema de indexar stdlib donde el primer elemento del rango es incluido y el último omitido. Por ejemplo, 0:10 las entradas son 0,...,9.

`loc` indexa incluye todos los indices. Por ejemplo, 0:10 regresa 0,...,10

# Agrupamiento

````python
import pandas as pd
````

````python
data1 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi',
                 'Gaurav', 'Anuj', 'Princi', 'Abhi'],
        'Age':[27, 24, 22, 32,
               33, 36, 27, 32],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',
                   'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd',
                         'B.Tech', 'B.com', 'Msc', 'MA']}
````

````python
df_data = pd.DataFrame(data1)
````

````python
df_data
````

````python
g_name = df_data.groupby('Name')
````

````python
print(df_data.groupby('Name').groups)
````

````python
g_name.last() ##Hace un ordenamiento por el grupo
````

````python
df_data.groupby(['Name','Qualification']).groups
````

````python
df_data.groupby(['Name']).sum()
````

````python
df_data.groupby(['Name'],sort=False).sum()
````
