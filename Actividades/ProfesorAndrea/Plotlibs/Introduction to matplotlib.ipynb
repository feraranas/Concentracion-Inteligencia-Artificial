# Pyplot API
Keeps track of the current figure and plotting area information and  plotting functions change that information.

We can get a reference to the current figure and axes using the following commands-

```plt.gcf ( ) # get current figure```

```plt.gca ( ) # get current axes```



````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x1 = np.linspace(0, 10, 100)
x1
````

````python
# create a plot figure
plt.figure()

# create the first of two panels and set current axis
plt.subplot(2, 1, 1)   # (rows, columns, panel number)
plt.plot(x1, np.sin(x1))


# create the second of two panels and set current axis
plt.subplot(2, 1, 2)   # (rows, columns, panel number)
plt.plot(x1, np.cos(x1))
````

````python
# get current figure information

print(plt.gcf())

# get current axis information

print(plt.gca())
````

# Object-Oriented API
This is available for more complex plotting situations. In the OO-API the plotting functions are methods of explicit Figure and Axes objects.

**Figure** is the top level container for all the plot elements. We can think of the **Figure** object as a box-like container containing one or more **Axes**.

The **Axes** represent an individual plot. The **Axes** object contain smaller objects such as axis, tick marks, lines, legends, title and text-boxes.

````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)


# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'c.-')
ax[1].plot(x1, np.cos(x1), 'b-');
````

````python
x = np.linspace(0,2*np.pi,100)
y = np.cos(x)
````

````python
fig = plt.figure()  # an empty figure with no Axes
fig, axs = plt.subplots(3, 3)  # a figure with a 3x3 grid of Axes
# a figure with one axes on the left, and two on the right:
axs[0, 0].plot(x, y)
axs[0,0].grid(True)

axs[0, 1].plot(y,x)
axs[0,2].scatter(x,y)
axs[0, 2].set_title('Axis [0, 2]')
axs[1,0].bar(x,y)
````

## Types of inputs to plotting functions

````python
np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
````

````python

fig, ax = plt.subplots(figsize=(5, 2.7))

ax.scatter('a', 'b', c='c', s='d', data=data) # Data is a dictionary
ax.set_xlabel('Entry a')
ax.set_ylabel('Entry b')
````
