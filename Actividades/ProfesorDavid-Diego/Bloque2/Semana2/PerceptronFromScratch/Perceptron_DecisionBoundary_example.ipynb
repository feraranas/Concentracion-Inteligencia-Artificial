# Preprocesing

## Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt

# plt.rcParams['text.usetex'] = True
```

## Generate Dataset - Logistic AND

```python
x1 = np.array([0, 0, 1, 1])
x2 = np.array([0, 1, 0, 1])
y  = np.array([0, 0, 0, 1])

w1 = 0.4
w2 = 0.4
theta = 0.5
```

## Set limits

```python
PAD = 0.5 # how much to "pad" around the min/max values. helps in setting bounds of plot

x1_min, x1_max = np.round(x1.min())-PAD, np.round(x1.max()+PAD)
x2_min, x2_max = np.round(x2.min())-PAD, np.round(x2.max()+PAD)

H = 0.1 # mesh stepsize
x1_axis_range = np.arange(x1_min, x1_max, H)
x2_axis_range = np.arange(x2_min, x2_max, H)
```

## Plot Meshgrid of the area

```python
# Create mesh
x1_mesh, x2_mesh = np.meshgrid(x1_axis_range, x2_axis_range)

# Get the shape of the grid for later
grid_shape = x1_mesh.shape

# Unpack the data to use the grid in the algorithm
xx1, xx2 = x1_mesh.ravel(), x2_mesh.ravel()

plt.scatter(x1_mesh, x2_mesh, s=0.5)
plt.show()
```

## Apply Activation function

```python
su = w1*xx1 + w2*xx2
yy = np.zeros_like(xx1)
idx = su > 0.5
yy[idx] = 1

yy = yy.astype(int)
```

# Making the Figure

### Plot the areas with its corresponding label

```python
redish = '#d73027'
orangeish = '#fc8d59'
yellowish = '#fee090'
blueish = '#4575b4'
colormap = np.array([redish,blueish,orangeish])

# make figure
plt.style.use('seaborn-whitegrid') # set style because it looks nice
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,6), dpi=150)

ax.scatter(xx1, xx2, c=colormap[yy], zorder=3, linewidths=0.7, s=10, edgecolor='k')
ax.scatter(x1, x2, c=colormap[y], zorder=3, linewidths=0.7, s=50, marker='X', edgecolor='k')

y_mesh = np.reshape(yy, grid_shape)
ax.contour(x1_mesh, x2_mesh, y_mesh,
           levels=3, linewidths=1,
           colors=[redish,blueish, blueish,orangeish,])

ax.plot([-0.5, 2.0], [0.0, 0.0], 'k')
ax.plot([0.0, 0.0], [-0.5, 2.0], 'k')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
```

### Save the image

```python
image_format = 'eps' # e.g .png, .svg, etc.
image_name = 'perceptron_decision_boundary.eps'

fig.savefig(image_name, format=image_format, dpi=1200)
```

### Plot again just with the final separating Hyperplane

```python
x1_bound = np.arange(x1_min, x1_max, 0.1)
x2_bound = theta/w2 - w1/w2*x1_bound

# make figure
plt.style.use('seaborn-whitegrid') # set style because it looks nice
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,6), dpi=150)

ax.scatter(x1, x2, c=colormap[y], zorder=3, linewidths=0.7, s=50, marker='X', edgecolor='k')

ax.plot(x1_bound, x2_bound, 'k')

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')

image_name = 'perceptron_decision_boundary_2.eps'
fig.savefig(image_name, format=image_format, dpi=1200)
```
