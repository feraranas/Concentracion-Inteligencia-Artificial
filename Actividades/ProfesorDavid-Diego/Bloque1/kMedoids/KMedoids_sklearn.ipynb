```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython import display
%matplotlib inline
```

kMedoid function

```
!pip install scikit-learn-extra
from sklearn_extra.cluster import KMedoids
```

```
dt = np.random.randint(0, 100, (100,2))
```

```
# Create a figure with a single subplot
# fig, ax = plt.subplots()

kmedoids = KMedoids(n_clusters=5, metric="manhattan", max_iter=5)
kmedoids.fit(dt)
```

```
clusters = kmedoids.cluster_centers_
medoids_final = dt[kmedoids.medoid_indices_]

print(medoids_final)
print(clusters)
```

```
plt.scatter(dt[:, 0], dt[:, 1], c = kmedoids.labels_)
plt.scatter(clusters[:, 0], clusters[:, 1], c = 'k', s = 200)
```
